```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_info_columns', 500)
pd.set_option('display.width', 1000)

#import player regular season
df = pd.read_excel('Player-Data-Set/NBA-Player-Statistics-season.xlsx', [0,1,2,3,4,5,6])

years = pd.read_csv('Player-Data-Set/NBA-playerlist.csv')
# years.head
years2 = years[['PERSON_ID', 'DISPLAY_FIRST_LAST', 'FROM_YEAR']]
# years2.head

data = pd.concat([df[0],df[1],df[2],df[3],df[4],df[5],df[6]])
#Order data by year asc
data = data.sort_values(by=['Year'])
#data for seasons
# data.head()

results = pd.read_excel ('Player-Data-Set/result.xlsx', [0,1,2,3,4,5,6])
df[0] = df[0].merge(results[0], left_on='TEAM_ID', right_on='TeamID', how='left')
df[1] = df[1].merge(results[1], left_on='TEAM_ID', right_on='TeamID', how='left')
df[2] = df[2].merge(results[2], left_on='TEAM_ID', right_on='TeamID', how='left')
df[3] = df[3].merge(results[3], left_on='TEAM_ID', right_on='TeamID', how='left')
df[4] = df[4].merge(results[4], left_on='TEAM_ID', right_on='TeamID', how='left')
df[5] = df[5].merge(results[5], left_on='TEAM_ID', right_on='TeamID', how='left')
df[6] = df[6].merge(results[6], left_on='TEAM_ID', right_on='TeamID', how='left')
data = pd.concat([df[0],df[1],df[2],df[3],df[4],df[5],df[6]])
#Order data by year asc
data = data.sort_values(by=['Year'])
#data for seasons
data.head()
```

```python
train = pd.concat([df[0],df[1],df[2],df[3],df[4]])
train.drop(train.columns[range(36,68)], axis=1, inplace=True)
train2=train
train2['PER'] = (train.FGM * 85.910 + train.STL * 53.897 + train.FG3M*51.757 + train.FTM * 46.845 + train.BLK * 39.190 + train.OREB * 39.190 + train.AST * 34.677 + train.DREB * 14.707 - train.PF * 17.174 - (train.FTA-train.FTM) * 20.091 - (train.FGA-train.FGM) * 39.190 - train.TOV * 53.897) / train.MIN
# train2.head
train2 = train2.merge(years2, left_on='PLAYER_ID', right_on='PERSON_ID', how='left')
train2.drop(['PERSON_ID'], axis=1, inplace=True)
train2['PLAYERYEAR'] = train2['PLAYER_ID'].map(str).str.cat(train2['Year'].map(str))
test = pd.concat([df[5],df[6]])
test.drop(test.columns[range(36,68)], axis=1, inplace=True)
test2=test
test2['PER'] = (test.FGM * 85.910 + test.STL * 53.897 + test.FG3M*51.757 + test.FTM * 46.845 + test.BLK * 39.190 + test.OREB * 39.190 + test.AST * 34.677 + test.DREB * 14.707 - test.PF * 17.174 - (test.FTA-test.FTM) * 20.091 - (test.FGA-test.FGM) * 39.190 - test.TOV * 53.897) / test.MIN
test2 = test2.merge(years2, left_on='PLAYER_ID', right_on='PERSON_ID', how='left')
test2['PLAYERYEAR'] = test2['PLAYER_ID'].map(str).str.cat(test2['Year'].map(str))
train2[train2['FROM_YEAR'].isnull()]
test2.drop(['PERSON_ID'], axis=1, inplace=True)

train2.to_csv("c:/Users/eve/OneDrive - Swinburne University/Documents/0_Project/Jupyter/train.csv", index = False)
test2.to_csv("c:/Users/eve/OneDrive - Swinburne University/Documents/0_Project/Jupyter/test.csv", index = False)
```
